````python
# 記得先 Import 正確的套件

import numpy as np
from numpy.random import default_rng
rng = default_rng()
````

### 3. 承上題，怎樣可以限制整數的範圍介於 m - n 之間？請將給定的 a 陣列當中的元素的範圍調整成 m - n 之間。


````python
m = 20
n = 40

# normal = rng.standard_normal((3,2)) 
# random = rng.random((3,2))
integers = rng.integers(m, n, size=(3,2))
print(integers)
````
